### Implementation details

For solving this environment we have used the Deep Q Learning Algorithm.

During training, the states are passed to the agent in agent.act() function.
This function uses an epsilon greedy approach to return either a random action or the action with the maximum 
Q-value fetched after passing through the Deep Q-Network (DQN). 

In each episode of training, the agent is passed the state of the environment, then it acts on it and decides an 
action, and then we get a Reward and new state. This tuple of (current_state, action, reward, next_state) is passed to 
the agent in agent.step() function. 

In agent.step() the agent stores each of these experience tuples in a Replay buffer. After every 4 such steps, if the Replay buffer size is more than the batch size (64), then we randomly sample 
the experience tuples and pass it through our DQN to help it learn using back-propagation. 

The DQN uses another concept called 'Fixed Target', for which it creates the target labels using another DQN with a fixed set of weights. These labels will be used to calculate the loss and do back-propagation. 
The Target DQN weights are then updated to the local DQN weights.

We repeat the process for 2000 episodes, and in each episode we run for 1000 iterations.This helps the agent learn the environment and achieve a good score. 

### Learning Algorithm

We have used Deep Q Learning Algorithm for training the agent. 
Within Deep Q Learning, we have used concepts like Replay buffer, and Fixed Targets for improving the algorithm. 

The training algorithm uses several hyperparameters, like epsilon start value, epsilon decay, minimum value of epsilon,
learning rate, discount factor. Even the Replay buffer size, number of steps to update the local DQN weights form hyperparameters. While copying the Local DQN weights to the Target DQN, we use a soft update parameter called TAU. This is also a hyperparameter. 

The DQN architecture itself is a very simple model consisting of an Input layer, followed by 3 Fully conneted layers. 
All the Fully connected layers have ReLU activations except for the last one. The number of neurons in each layer of the DQN is 
also a hyperparameter.

### Ideas for Future work

The model's performance can be improved if we use a deeper Neural network with more layers. Apart from this, the Deep Q Learning algorithm itself 
can be improved even more using techniques like Double DQN, Prioritized experience replay, Dueling DQN, Learning from multi-step bootstrap targets, Distributed DQN, Noisy DQN, or a combination of all above called Rainbow.
